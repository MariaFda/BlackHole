```python
from scipy.special import erf
x= linspace(0.0,1.0, 10)
plot(x, erf(x))
```

```python
h=0.001
n_points = int((1.0+h)/h)
t = zeros(n_points)
r = zeros(n_points)
v = zeros(n_points)
m = zeros(n_points)
```

```python
t[0] = 0 
r[0] = 0.0
v[0] = 500.0
m[0] = 10.0E6 
```

```python
G_GRAV = 4.9E-6 #unidades desconocidas
SIGMA = 400.0
LOG_LAMBDA = 12.0
M_TOT = 1.0E10
RHO = 1.23*10**7
M_BH= 10.0E6
C=1.4*10**4
def func_r_prime(t, r, v):
    return v

def func_m_prime(t,r,v):
    part_a=pi*exp(3.0/2.0)*RHO*G_GRAV**2*M_BH**2*C**-3
    part_b= (1+(v**2/C**2))**3/2
    return part_a/part_b
    

def func_v_prime(t, r, v):
    part_a =  G_GRAV*(4.0*pi/3.0) * RHO * r
    part_b = 4.0*pi*G_GRAV**2 * LOG_LAMBDA * RHO * M_TOT / v**2
    ratio = v/(sqrt(2.0)* SIGMA)
    part_c = erf(ratio) - 2.0*ratio*exp(-(ratio**2))/sqrt(pi)
    # part_d= v
    return  -part_a - (part_b * part_c)



for i in range(1,n_points):
    #get the first derivatives
    r_prime = func_r_prime(t[i-1],r[i-1], v[i-1])
    v_prime = func_v_prime(t[i-1],r[i-1], v[i-1])
    m_prime = func_m_prime(t[i-1],r[i-1], v[i-1])
    
    #from this estimation move to the middle point
    t_middle = t[i-1] + (h/2.0)
    r_middle = r[i-1] + (h/2.0) * r_prime
    v_middle = v[i-1] + (h/2.0) * v_prime 
    m_middle = m[i-1] + (h/2.0) * m_prime 
    
    #compute the derivatives at the middle point
    r_middle_prime = func_r_prime(t_middle, r_middle, v_middle)
    v_middle_prime = func_v_prime(t_middle, r_middle, v_middle)
    m_middle_prime = func_m_prime(t_middle, r_middle, v_middle)
    
    t[i] = t[i-1] + h
    r[i] = r[i-1] + h * r_middle_prime 
    v[i] = v[i-1] + h * v_middle_prime
    m[i] = m[i-1] + h * m_middle_prime
```

```python
plot(t,r)
#Gráfica tiempo(giga-years) vs posición(kiloparsec) 
```
